```python
df.dropna(inplace=True)
```

```python
x=df.iloc[:,1:-1]
y=df.iloc[:,-1]
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3, random_state=42)
print(x_train)
```

```python
print(x_test)
```

```python
import seaborn as sns

# Assuming you already have the dataset loaded as a pandas DataFrame

# Plot box plot for all features
plt.figure(figsize=(12, 8))
sns.boxplot(data=x_train)
plt.title("Box Plot for Features")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming you already have the dataset loaded as a pandas DataFrame

# Plot histograms for all features
x_train.hist(figsize=(20, 20))
plt.suptitle("Histograms of Features", y=1.02)
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

model = GaussianNB()
model.fit(x_train, y_train)

predictions = model.predict(x_test)

accuracy = accuracy_score(y_test, predictions)
print("Model Accuracy:", accuracy)

cm_nb = confusion_matrix(y_test, predictions)
print("Confusion Matrix - Naive Bayes:")
print(cm_nb)

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_train)

predictions = model.predict(x_test)

accuracy = accuracy_score(y_test, predictions)
print("Model Accuracy:", accuracy)

cm = confusion_matrix(y_test, predictions)
print("Confusion Matrix - Random Forest:")
print(cm)

```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score


model = DecisionTreeClassifier(random_state=42)
model.fit(x_train, y_train)

predictions = model.predict(x_test)

accuracy = accuracy_score(y_test, predictions)
print("Model Accuracy:", accuracy)


cm= confusion_matrix(y_test, predictions)
print("Confusion Matrix - Decision tree:")
print(cm)


```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

model_knn = KNeighborsClassifier(n_neighbors=5)
model_knn.fit(x_train, y_train)

predictions = model_knn.predict(x_test)

accuracy = accuracy_score(y_test, predictions)
print("Model Accuracy:", accuracy)

cm = confusion_matrix(y_test, predictions)
print("Confusion Matrix - KNN")
print(cm)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

model = SVC(kernel='linear', random_state=42)  # You can try different kernels like 'linear', 'rbf', 'poly', etc.
model.fit(x_train, y_train)

# Test the model on the test set
predictions = model.predict(x_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, predictions)
print("Model Accuracy:", accuracy)

cm = confusion_matrix(y_test, predictions)
print("Confusion Matrix - SVM:")
print(cm)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score

# Assuming you already have x_train, x_test, y_train, and y_test datasets

# Create and train the Gradient Boosting classifier
model_gbm = GradientBoostingClassifier(n_estimators=100, random_state=42)  # You can adjust the number of estimators as needed
model_gbm.fit(x_train, y_train)

# Test the model on the test set
predictions_gbm = model_gbm.predict(x_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, predictions_gbm)
print("Model Accuracy:", accuracy)

cm = confusion_matrix(y_test, predictions_gbm)
print("Confusion Matrix - Gradient Boosting Machines :")
print(cm)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier


# Assuming you already have x_train, x_test, y_train, and y_test datasets

# Create and train the Gaussian Naive Bayes model
model_nb = GaussianNB()
model_nb.fit(x_train, y_train)
predictions_nb = model_nb.predict(x_test)
cm_nb = confusion_matrix(y_test, predictions_nb)

# Create and train the Random Forest classifier
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf.fit(x_train, y_train)
predictions_rf = model_rf.predict(x_test)
cm_rf = confusion_matrix(y_test, predictions_rf)

# Create and train the Decision Tree classifier
model_dt = DecisionTreeClassifier(random_state=42)
model_dt.fit(x_train, y_train)
predictions_dt = model_dt.predict(x_test)
cm_dt = confusion_matrix(y_test, predictions_dt)

# Create and train the knn
model_knn = KNeighborsClassifier(n_neighbors=5)
model_knn.fit(x_train, y_train)
predictions = model_knn.predict(x_test)
cm = confusion_matrix(y_test, predictions)


# Plot the heatmaps
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Heatmap for Naive Bayes
sns.heatmap(cm_nb, annot=True, cmap='Blues', fmt='d', ax=axes[0, 0])
axes[0, 0].set_title("Confusion Matrix - Naive Bayes")

# Heatmap for Random Forest
sns.heatmap(cm_rf, annot=True, cmap='Greens', fmt='d', ax=axes[0, 1])
axes[0, 1].set_title("Confusion Matrix - Random Forest")

# Heatmap for Decision Tree
sns.heatmap(cm_dt, annot=True, cmap='Oranges', fmt='d', ax=axes[1, 0])
axes[1, 0].set_title("Confusion Matrix - Decision Tree")

# Heatmap for KNN
sns.heatmap(cm, annot=True, cmap='Reds', fmt='d', ax=axes[1, 1])
axes[1, 1].set_title("Confusion Matrix - KNN")

plt.tight_layout()
plt.show()
```
